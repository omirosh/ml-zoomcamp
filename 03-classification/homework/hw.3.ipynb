```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/data.csv")
```

```python
df
```

```python
df = df[
    [
        "Make",
        "Model",
        "Year",
        "Engine HP",
        "Engine Cylinders",
        "Transmission Type",
        "Vehicle Style",
        "highway MPG",
        "city mpg",
        "MSRP",
    ]
]
```

```python
df.columns = df.columns.str.replace(" ", "_").str.lower()
```

```python
df
```

```python
df.info()
```

```python
df[["engine_hp", "engine_cylinders"]] = df[["engine_hp", "engine_cylinders"]].fillna(0)
```

```python
df.rename(columns={"msrp": "price"}, inplace=True)
```

```python
df.transmission_type.describe()  # Q1: AUTOMATIC
```

```python
import seaborn as sns
```

```python
sns.heatmap(
    df.select_dtypes(include=np.number).corr(), annot=True
)  # Q2: highway_mpg and engine_cylinders
```

```python
mean_price = df.price.mean()
df["above_average"] = np.where(df.price > mean_price, 1, 0)
```

```python
from sklearn.model_selection import train_test_split

target = "above_average"
seed = 42
X_train, X_test, y_train, y_test = train_test_split(
    df, df[target], test_size=0.2, random_state=seed  # .drop(target, axis=1)
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.25, random_state=seed
)
```

```python
from sklearn.metrics import mutual_info_score

for col in df.select_dtypes(include="object").columns:
    print(
        col, round(mutual_info_score(X_train[target], X_train[col]), 2)
    )  # Q3: transmission_type
X_train.drop(target, axis=1, inplace=True)
X_val.drop(target, axis=1, inplace=True)
X_test.drop(target, axis=1, inplace=True)
```

```python
X_train.drop("price", axis=1, inplace=True)
X_val.drop("price", axis=1, inplace=True)
X_test.drop("price", axis=1, inplace=True)
```

```python
from sklearn.feature_extraction import DictVectorizer

train_dicts = X_train.to_dict(orient="records")
val_dicts = X_val.to_dict(orient="records")
test_dicts = X_test.to_dict(orient="records")

dv = DictVectorizer(sparse=False)
X_train_vect = dv.fit_transform(train_dicts)
X_val_vect = dv.transform(val_dicts)
X_test_vect = dv.transform(test_dicts)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver="liblinear", C=10, max_iter=1000, random_state=42)
```

```python
model.fit(X_train_vect, y_train)
```

```python
model.coef_
```

```python
from sklearn.metrics import accuracy_score
```

```python
pred_val_hard = model.predict(X_val_vect)
```

```python
pred_val = model.predict_proba(X_val_vect)[:, 1]
```

```python
decision = (pred_val > 0.5).astype("int")
```

```python
round(accuracy_score(y_val, pred_val_hard), 2)
```

```python
original_accuracy = round((y_val == decision).mean(), 2)  # Q4: 0.95
original_accuracy
```

```python
cols = list(X_train.columns)
```

```python
for c in cols:
    cols = list(X_train.columns)
    cols.remove(c)
    X_train_c = X_train.copy()
    X_train_c = X_train_c[cols]
    X_val_c = X_val.copy()
    X_val_c = X_val_c[cols]
    train_dicts = X_train_c.to_dict(orient="records")
    val_dicts = X_val_c.to_dict(orient="records")

    X_train_vect = dv.transform(train_dicts)
    X_val_vect = dv.transform(val_dicts)
    model.fit(X_train_vect, y_train)
    pred_val = model.predict(X_val_vect)
    print(
        "Removing ",
        c,
        " accuracy = ",
        round(accuracy_score(y_val, pred_val), 3),
        "difference: ",
        round(original_accuracy - round(accuracy_score(y_val, pred_val), 3), 3),
    )
    cols.pop()
    # Q5: year
```

```python
target_new = "price"
df[target_new] = np.log1p(df[target_new])
print(df.columns)
df.drop("above_average", axis=1, inplace=True)
seed = 42
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(target_new, axis=1), df[target_new], test_size=0.2, random_state=seed
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.25, random_state=seed
)
```

```python
from sklearn.linear_model import Ridge
```

```python
train_dicts = X_train.to_dict(orient="records")
val_dicts = X_val.to_dict(orient="records")
test_dicts = X_test.to_dict(orient="records")

dv = DictVectorizer(sparse=True)
X_train_vect = dv.fit_transform(train_dicts)
X_val_vect = dv.transform(val_dicts)
X_test_vect = dv.transform(test_dicts)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
for a in [0, 0.01, 0.1, 1, 10]:
    model = Ridge(alpha=a, random_state=seed, solver="sag")
    model.fit(X_train_vect, y_train)
    y_pred = model.predict(X_val_vect)
    rmse = mean_squared_error(y_val, y_pred, squared=False)
    print("alpha = ", a, "rmse = ", round(rmse, 3))
    # Q6: alpha = 0
```
